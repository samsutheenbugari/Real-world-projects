## IMPORT LIBRARIES AND DATA

```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
```

```python
movie_data=pd.read_csv("movies.csv")
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
movie_data.columns.tolist()
```

```python
movie_data.isnull().sum()
```

```python
# selecting the relavent feature
features=['genres','keywords','tagline','cast','director']

data=pd.DataFrame(data=movie_data,columns=features)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Replace the null values
for feature in features:
    data[feature]=data[feature].fillna("")
    
```

```python
data.isnull().sum()
```

```python
#combining all the 5 selected features

combined_data=data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
```

```python
combined_data
```

```python
# converting the TEXT data into (NUMERICAL)  VECTOR data

vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_data)
```

```python
print(feature_vector)
```

## Cosine Similarity

```python
# getting the similarity scores using cosine similarity

similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
#getting the input movie fro user
movie_name=input('Enter Your Favourite Movie name:')
```

```python
# create a list with all the movies given in the data set

movies_list=movie_data['title'].tolist()
```

```python
movies_list
```

```python
# finding the close match of the movie name given by user

close_match=difflib.get_close_matches(movie_name,movies_list)
close_match
```

```python
close_match=close_match[0]
close_match
```

```python
#finding the index of the movie in title

index_of_the_movie=movie_data[movie_data.title == close_match]['index'].values[0]
index_of_the_movie
```

```python
#getting the list of similar movie

similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on similarity score

sorted_movies=sorted(similarity_score,key= lambda x:x[1], reverse=True)
sorted_movies
```

```python
# print the nsme of similar movie based on index
print('Movies Suggested For You:\n')
i=1

for movie in sorted_movies:
    index=movie[0]
    title_from_index=movie_data[movie_data.index==index]['title'].values[0]
    if(i<30):
        print(i,'.', title_from_index)
        i+=1
```

# AUTOMATED SYSTEM

```python
movie_name=input('Enter Your Favourite Movie name:')


movies_list=movie_data['title'].tolist()

close_match=difflib.get_close_matches(movie_name,movies_list)

close_match=close_match[0]

index_of_the_movie=movie_data[movie_data.title == close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_movies=sorted(similarity_score,key= lambda x:x[1], reverse=True)

print('Movies Suggested For You:\n')
i=1

for movie in sorted_movies:
    index=movie[0]
    title_from_index=movie_data[movie_data.index==index]['title'].values[0]
    if(i<30):
        print(i,'.', title_from_index)
        i+=1
```
