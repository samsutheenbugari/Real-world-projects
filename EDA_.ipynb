##### 10.CURRENCY CONVERSION

```python
def change(row, region):
    if region == 'USA':
        return round(row * 0.012, 2)
    elif region == 'UAE':
        return round(row * 0.043, 2)
    elif region == 'UK':
        return round(row * 0.0092, 2)
    elif region == 'CAN':
        return round(row * 0.017, 2)
    else:
        return row  


df3['Lpp'] = df3.apply(lambda row: change(row['Lpp'], row['region']), axis=1)
```
